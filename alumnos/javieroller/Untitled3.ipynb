```python

import random
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
import numpy as np
import sys
%matplotlib inline



#zeta = 1.6875

def rad_1s(r1,r2):
    zeta = 1.6875
    fi1122 = (1/np.pi*4*np.sqrt(2))*(zeta**3)*(2-zeta*r2)*np.exp(-zeta*(r1 + 0.5*r2))
    fi1221 = (1/np.pi*4*np.sqrt(2))*(zeta**3)*(2-zeta*r1)*np.exp(-zeta*(r2 + 0.5*r1))
    
    fdr = (1/np.sqrt(2))*(fi1122 - fi1221)*(1/np.sqrt(2))*(fi1122 - fi1221)*(r1**2)*(r2**2)
    
    return fdr


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
r1 = r2 = np.arange(0.0, 4.0, 0.005)

R1, R2 = np.meshgrid(r1, r2)
zs = np.array([rad_1s(r1,r2) for r1,r2 in zip(np.ravel(R1), np.ravel(R2))])
Z = zs.reshape(R1.shape)

ax.plot_surface(R1, R2, Z)

ax.set_xlabel('r1 ')
ax.set_ylabel('r2 ')
ax.set_zlabel('Z ')

plt.show()

```
