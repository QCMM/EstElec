```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
import numpy as np
import sys
%matplotlib inline

zeta = 1.6875

def r(x,y,z):
    return np.sqrt(x**2+y**2+z**2)

def cart_1s(x1,y1,z1, x2,y2,z2,zeta):
    
    fi1122 = (1/np.pi*4*np.sqrt(2))*(zeta**3)*(2-zeta*r(x2,y2,z2))*np.exp(-zeta*(r(x1,y1,z1)+0.5*r(x2,y2,z2)))
    fi1221 = (1/np.pi*4*np.sqrt(2))*(zeta**3)*(2-zeta*r(x1,y1,z1))*np.exp(-zeta*(r(x2,y2,z2)+0.5*r(x1,y1,z1)))
    
    fdr = (1/np.sqrt(2))*(fi1122 - fi1221)*(1/np.sqrt(2))*(fi1122 - fi1221)*((r(x1,y1,z1))**2)*((r(x2,y2,z2))**2)
    
    return fdr

x1 = np.linspace(-4,4,2000)
y1 = np.linspace(-4,4,2000)
z1 = 0.0
X1,Y1 = np.meshgrid(x1, y1)

x2 = 0.5
z2 = 0.0
y2 = 0.0

def plot_3D_cart(X,Y,z,x1,y1,z1,ze,func):
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    surf = ax.plot_surface(X, Y, func(X,Y,z,x1,y1,z1,ze), cmap=cm.coolwarm,
                              linewidth=0, antialiased=False)

    # Customize the z axis.
    ax.zaxis.set_major_locator(LinearLocator(10))
    ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

    # Add a color bar which maps values to colors.
    fig.colorbar(surf, shrink=0.5, aspect=5)
    
plot_3D_cart(X1,Y1,z1,x2,y2,z2,zeta,cart_1s)    
```
