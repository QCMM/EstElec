```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import random
```

```python
zeta = 1.69
```

```python
def fun(x, y):
    return (1/np.sqrt(2))**2*(zeta)**6 *(1/(4*np.pi*np.sqrt(2)))**2*((2-zeta*x)*np.exp(-zeta*(x+y/2))-(2-zeta*y)*np.exp(-zeta*(y+x/2)))**2 * x**2 *y**2
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = y = np.arange(0.0, 1.0, 0.05)
X, Y = np.meshgrid(x, y)
zs = np.array([fun(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z)

ax.set_xlabel('R1')
ax.set_ylabel('R2')
ax.set_zlabel('distribucion radial')

plt.show()
```
