```python
def angles(coords):
    for i in range(0,int(n_a)):
        for j in range(i+1,int(n_a)):
            R_ij = math.sqrt((coords[1][i]- coords[1][j])**2 + (coords[2][i]- coords[2][j])**2 + (coords[3][i]- coords[3][j])**2)
            for k in range(j+1,int(n_a)):
                R_jk = math.sqrt((coords[1][j]- coords[1][k])**2 + (coords[2][j]- coords[2][k])**2 + (coords[3][j]- coords[3][k])**2)
                teta_jik_rad = np.arccos((-(coords[1][j]-coords[1][i])/(R_ij))*(-(coords[1][j]-coords[1][k])/(R_jk)) + (-(coords[2][j]-coords[2][i])/(R_ij))*(-(coords[2][j]-coords[2][k])/(R_jk))+(-(coords[3][j]-coords[3][i])/(R_ij))*(-(coords[3][j]-coords[3][k])/(R_jk)))
                teta_jik = teta_jik_rad * 180 / math.pi
                print(str(i)+ " "+str(j)+ " "+str(k))
                print(teta_jik)
                print("-----")
                    
    return(teta_jik)
```

```python
xyz = get_xyz(lines)
n_a = lines[0]

angles = angles(xyz)
print(angles)
```
