```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
import numpy as np
import math
%matplotlib inline

x = np.linspace(-4,4,1000)
y = np.linspace(-4,4,1000)
z = np.linspace(-4,4,1000)
r = np.linspace(0,8,1000)
X,Y = np.meshgrid(x, y)
```

```python
def psi_r(r):
    return math.sqrt(1/math.pi)*np.exp(-r)
```

```python
plt.plot(r,psi_r(r))
plt.xlabel("r/a_0")
plt.ylabel("Psi_100/a_0")
```

```python
def p_r(r):
    return 4*1/np.sqrt(np.pi)*np.exp(-r)*
```

```python
plt.plot(r,p_r(r))
plt.xlabel("r/a_0")
plt.ylabel("Psi_100/a_0")
```

```python
def psi_x_100(x):
    return math.sqrt(1/math.pi)*np.exp(-abs(x))

def psi_xy_100(x,y):
    return math.sqrt(1/math.pi)*np.exp(-math.sqrt(x**2 + y**2))
```

```python
plt.plot(x,psi_x_100(x)**2)
plt.xlabel("X/a_0")
plt.ylabel("Psi_100/a_0")
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

# Plot the surface.
f2 = np.vectorize(psi_xy_100)
surf = ax.plot_surface(X, Y, f2(X,Y)**2, cmap=cm.coolwarm, linewidth=0, antialiased=False)

# Customize the z axis.
#ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)
```
