```python
import sys, math
import numpy as np
```

```python
f = open("../acetaldehido.xyz")
lines = f.readlines()
print(lines[1])

def get_xyz(lines):
    x_coords = []
    y_coords = []
    z_coords = []
    n_atoms = None
    atom_list = []
    n_atoms = lines[0]
    for i in range(1,len(lines)):
        atom = lines[i].split()[0]
        x    = lines[i].split()[1]
        y    = lines[i].split()[2]
        z    = lines[i].split()[3]
        x_coords.append(float(x))
        y_coords.append(float(y))
        z_coords.append(float(z))
        atom_list.append(int(atom))
        
    return(atom_list, x_coords, y_coords, z_coords)


```

```python
def angles(coords):
    for i in range(0,int(n_a)):
        for j in range(i+1,int(n_a)):
            R_ij = math.sqrt((coords[1][i]- coords[1][j])**2 + (coords[2][i]- coords[2][j])**2 + (coords[3][i]- coords[3][j])**2)
            for k in range(j+1,int(n_a)):
                R_jk = math.sqrt((coords[1][j]- coords[1][k])**2 + (coords[2][j]- coords[2][k])**2 + (coords[3][j]- coords[3][k])**2)
                teta_ijk_rad = np.arccos((-(coords[1][j]-coords[1][i])/(R_ij))*(-(coords[1][j]-coords[1][k])/(R_jk)) + (-(coords[2][j]-coords[2][i])/(R_ij))*(-(coords[2][j]-coords[2][k])/(R_jk))+(-(coords[3][j]-coords[3][i])/(R_ij))*(-(coords[3][j]-coords[3][k])/(R_jk)))
                teta_ijk = teta_ijk_rad * 180 / math.pi
                print(str(i)+ " "+str(j)+ " "+str(k))
                print(teta_ijk)
                print("-----")
                    
    return(teta_ijk)
```

```python
xyz = get_xyz(lines)
n_a = lines[0]

angles = angles(xyz)
print(angles)
```
